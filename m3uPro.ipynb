```python
import os
import requests

def get_proxy_list(choice):
    urls = {
        1: "https://api.proxyscrape.com/v2/?request=displayproxies&protocol=http&timeout=10000&country=all&ssl=all&anonymity=all",
        2: "https://raw.githubusercontent.com/proxy4parsing/proxy-list/main/http.txt",
        3: "https://raw.githubusercontent.com/TheSpeedX/PROXY-List/master/http.txt",
        4: "https://api.openproxylist.xyz/http.txt",
        5: "https://raw.githubusercontent.com/ErcinDedeoglu/proxies/main/proxies/socks4.txt",
        6: "https://api.openproxylist.xyz/socks4.txt",
        7: "https://raw.githubusercontent.com/ErcinDedeoglu/proxies/main/proxies/socks5.txt",
        8: "https://api.openproxylist.xyz/socks5.txt",
        9: "https://api.proxyscrape.com/v2/?request=displayproxies&protocol=all&timeout=750&country=US",
        10: "https://api.proxyscrape.com/v2/?request=displayproxies&protocol=all&timeout=1000&country=CA",
        11: "https://api.proxyscrape.com/v2/?request=displayproxies&protocol=all&timeout=1000&country=IT",
        12: "https://api.proxyscrape.com/v2/?request=displayproxies&protocol=all&timeout=1000&country=FR",
        13: "https://api.proxyscrape.com/v2/?request=displayproxies&protocol=all&timeout=1500&country=DE",
        14: "https://api.proxyscrape.com/v2/?request=displayproxies&protocol=all&timeout=1250&country=CN",
        15: "https://proxy.webshare.io/api/v2/proxy/list/download/zrqapzfmrpmrqdbjqultdqlenaqpqjjgdhpwvdgi/-/any/username/direct/-/"
    }

    filenames = {
        1: 'HTTP¹.txt',
        2: 'HTTP².txt',
        3: 'HTTP³.txt',
        4: 'HTTP⁴.txt',
        5: 'SOCKS-4¹.txt',
        6: 'SOCKS-4².txt',
        7: 'SOCKS-5¹.txt',
        8: 'SOCKS-5².txt',
        9: 'USA.txt',
        10: 'CANADA.txt',
        11: 'ITALY.txt',
        12: 'FRANCE.txt',
        13: 'GERMANY.txt',
        14: 'CHINA.txt'
    }

    url = urls.get(choice)
    filename = filenames.get(choice)
    return url, filename

def download_proxy(url, save_path):
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'}
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        with open(save_path, 'w') as file:
            file.write(response.text)
        print(f'Proxies downloaded and saved to {save_path}')
    else:
        print(f'Failed to download proxies. Status code: {response.status_code}')

def main():
    print('''Choose a proxy list to download:
     1. HTTP¹
     2. HTTP²
     3. HTTP³
     4. HTTP⁴
     5. SOCKS4¹
     6. SOCKS4²
     7. SOCKS5¹
     8. SOCKS5²
     9. USA
    10. CANADA
    11. ITALY
    12. FRANCE
    13. GERMANY
    14. CHINA
     0. EXIT
    ''')

    choice = int(input("Enter the number of your choice: "))
    if choice == 0:
        print("Exiting...")
        return

    url, filename = get_proxy_list(choice)
    if not url:
        print("Invalid choice. Exiting...")
        return

    save_dir = input("Enter the directory to save the proxies: ").strip()
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    save_path = os.path.join(save_dir, filename)
    download_proxy(url, save_path)

if __name__ == "__main__":
    main()
```

```python
import requests
import concurrent.futures
import os

def check_proxy(proxy):
    url = "http://www.google.com"  # URL to test the proxy against
    proxies = {
        "http": f"http://{proxy}",
        "https": f"http://{proxy}",
    }
    try:
        response = requests.get(url, proxies=proxies, timeout=5)
        if response.status_code == 200:
            return proxy, True
    except:
        return proxy, False
    return proxy, False

def load_proxies(file_path):
    with open(file_path, 'r') as file:
        proxies = file.readlines()
    return [proxy.strip() for proxy in proxies]

def save_alive_proxies(alive_proxies, save_path):
    with open(save_path, 'w') as file:
        for proxy in alive_proxies:
            file.write(f"{proxy}\n")

def main():
    file_path = input("Enter the path of the proxy list file: ").strip()
    save_path = input("Enter the path to save the alive proxies: ").strip()

    if not os.path.exists(os.path.dirname(save_path)):
        os.makedirs(os.path.dirname(save_path))

    proxies = load_proxies(file_path)
    alive_proxies = []

    with concurrent.futures.ThreadPoolExecutor(max_workers=20) as executor:
        future_to_proxy = {executor.submit(check_proxy, proxy): proxy for proxy in proxies}
        for future in concurrent.futures.as_completed(future_to_proxy):
            proxy, is_alive = future.result()
            if is_alive:
                print(f"Proxy {proxy} is alive")
                alive_proxies.append(proxy)
            else:
                print(f"Proxy {proxy} is dead")

    save_alive_proxies(alive_proxies, save_path)
    print(f"Alive proxies saved to {save_path}")

if __name__ == "__main__":
    main()
```
